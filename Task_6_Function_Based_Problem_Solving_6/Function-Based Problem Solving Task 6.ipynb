## Problem Statement 1: Smart Parking Lot Management System

```python
def parking_lot_manager(logs):
    # Maximum number of vehicles the parking lot can hold
    parking_capacity = 50

    # Counters for vehicle entries and exits
    in_count = 0
    out_count = 0

    # Processing each log entry
    for log in logs:
        if log == 'IN':
            # Vehicle entered the parking lot
            in_count += 1
        elif log == 'OUT':
            # Vehicle exited the parking lot
            out_count += 1
        else:
            # Invalid input (not IN or OUT)
            print('Invalid Entry')

    # Calculating currently parked vehicles
    parked = in_count - out_count

    # Checking parking availability
    if parked < parking_capacity:
        status = 'Available'
    else:
        status = 'Not Available'

    # Returning number of parked vehicles and parking status
    return parked, status

# Number of IN/OUT logs to be entered
entries = int(input('Enter the number of entry and exit logs: '))

# List to store all log entries
log_list = []

# Collecting each log from the user
for entry in range(entries):
    log = input('Enter Vehicle IN/OUT: ').upper()
    log_list.append(log)

# Calling the parking lot manager function
parked, status = parking_lot_manager(log_list)

# Displaying final results
print('Currently Parked Vehicles:', parked)
print('Parking Status:', status)
```

## Problem Statement 2: Online Food Delivery Time Estimator

```python
def delivery_time_estimator(distance, traffic, weather):
    # assuming base speed as 20 km/hr and converting travel time to minutes
    base_time = (distance / 20) * 60

    # Determining traffic delay
    if traffic == "heavy":
        traffic_delay = 20
    elif traffic == "medium":
        traffic_delay = 10
    else:  
        traffic_delay = 0 #No traffic

    # Determining weather delay
    if weather == "stormy":
        weather_delay = 25
    elif weather == "rainy":
        weather_delay = 15
    else:  
        weather_delay = 0 #clear weather

    # Total estimated time
    total_time = base_time + traffic_delay + weather_delay

    # Returning rounded ETA
    return round(total_time)

# User input distance in kilometers
distance = float(input("Distance (km):"))

# User input traffic level
traffic = input("Traffic Level (Low/Medium/Heavy):").lower()

# User input weather condition
weather = input("Weather (Clear/Rainy/Stormy):").lower()

# Calculating estimated delivery time
eta = delivery_time_estimator(distance, traffic, weather)

# Displaying the result
print(f"Estimated Delivery Time: {eta} minutes")
```

## Problem Statement 3: Movie Theatre Seat Occupancy Analyzer

```python
def seat_occupancy_analyzer(total_seats, booked_count):
    # Calculating the occupancy percentage
    occupancy = (booked_count / total_seats) * 100

    # Deciding the show status
    if occupancy == 100:
        status = "Housefull"
    elif occupancy >= 70:
        status = "Almost Full"
    else:
        status = "Seats Available"

    # Suggesting opening extra shows
    if occupancy >= 80:
        suggestion = "Consider opening additional shows"
    else:
        suggestion = "No additional shows needed"

    # Printing the results
    print(f"Occupancy: {occupancy:.0f}%")
    print(f"Show Status: {status}")
    print(f"Suggestion: {suggestion}")

# Total number of seats
total_seats = int(input("Enter total seats: "))

# Number of booked seats
booked_count = int(input("Enter number of booked seats: "))

# Analyzing the seat occupancy
seat_occupancy_analyzer(total_seats, booked_count)
```

## Problem Statement 4: Cloud Server Load Classification System

```python
def server_load_classifier(readings):
    total = 0

    # adding all CPU readings
    for reading in readings:
        total += reading

    # calculating average CPU load
    avg = total / len(readings)

    # determining server status based on average load
    if avg < 50:
        status = 'Normal'
    elif avg < 80:
        status = 'Warning'
    else:
        status = 'Critical'

    # returning average and status
    return avg, status


# taking number of readings
num_readings = int(input('Enter the total number of readings: '))

readings = []

# collecting CPU readings
for r in range(num_readings):
    load = int(input(f'Enter the CPU reading {r+1}: '))
    readings.append(load)

# calling function and receiving results
avg, status = server_load_classifier(readings)

# printing final output
print(f'Average CPU Load: {avg:.0f}%')
print('Server Status:', status)
```

## Problem Statement 5: Smart Classroom Resource Usage Monitor

```python
def resource_usage_monitor(usage):
    # List to store overused resources
    overused = []

    # Threshold for overuse (in hours)
    limit = 8

    # Checking each resource usage
    for resource, hours in usage.items():
        if hours > limit:
            overused.append(resource)

    # Printing results
    if overused:
        print("Overused Resources:", ", ".join(overused))
        print("Energy Alert: Yes")
    else:
        print("Overused Resources: None")
        print("Energy Alert: No")

#empty dictionary to store resources and usage hours
resources = {}

#number of resources
n = int(input("Enter number of resources: "))

#user input of each resource name and usage hours
for i in range(n):
    name = input(f"Enter resource {i+1} name: ")
    hours = int(input(f"Enter usage hours for {name}: "))
    resources[name] = hours

# Analyzing resource usage
resource_usage_monitor(resources)
```

## Problem Statement 6: Online Event Registration Capacity Controller

```python
def event_capacity_controller(capacity, registrations):
    # Check if registrations exceed capacity
    if registrations > capacity:
        confirmed = capacity
        waitlisted = registrations - capacity
        status = "Closed"
    else:
        confirmed = registrations
        waitlisted = 0
        status = "Open"

    # Returning the results
    return confirmed, waitlisted, status

#user input for capacity and number of registrations
capacity = int(input("Enter event capacity: "))
registrations = int(input("Enter total registrations: "))

# Function calling
confirmed, waitlisted, status = event_capacity_controller(capacity, registrations)

# Printing the output
print("Confirmed Registrations:", confirmed)
print("Waitlisted Users:", waitlisted)
print("Registration Status:", status)
```
