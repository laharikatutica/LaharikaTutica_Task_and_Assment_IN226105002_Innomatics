## Problem 1: Unique Words in a Sentence

```python
#taking a user input sentence
sentence = input('Enter sentence: ')

#splitting the sentence into a list of words
words = sentence.split()  

#converting the list into a set
unique_words = set(words) 

#printing the Unique words count and the set of unique words
print('Unique words count:',len(unique_words))
print('Unique words:',unique_words)
```

## Problem 2: Highest Salary from Employee Data

```python
#creating an empty dictionary to store keys representing employee names and values representing employee salaries
employees = {}

#taking input until the user decides to stop
while True:
    # asking the user to enter an employee name
    name = input("Enter employee name (press Enter to stop): ") #if pressed enter, the loop will end
    
    if name == "":
        break # exit the loop if pressed enter

    # taking salary input for the entered employee
    salary = int(input("Enter salary: "))
    # storing the employee name and salary in the dictionary
    employees[name] = salary

#print(employees)

#assuming the first employee salary as the highest
highest_name,highest_salary = next(iter(employees.items()))

#iterating through each employee salary to get the highest salary
for name,salary in employees.items():
    if salary > highest_salary:
        highest_salary = salary
        highest_name = name

#printing the highest salary with the employee name
print(f'Highest Salary: {highest_name} - {highest_salary}')
```

## Problem 3: Find Maximum and Minimum Values

```python
#taking user input for list of integers(numeric values)
numbers = eval(input('Enter list of numbers: '))

#assuming the minimum and maximum to be the first integer values of the list
min_num = numbers[0]
max_num = numbers[0]

#iterating through each number of the list
for num in numbers:
    #checking for minimum number
    if num < min_num:
        min_num = num
    
    #checking for maximum number
    if num > max_num:
        max_num = num
        
#printing the minimum and maximum numbers from the list and the list itself
print('List:',numbers)
print('Max:',max_num)
print('Min:',min_num)
```

## Problem 4: Count Products Above a Price Threshold

```python
#taking user input for a list of prices of the products
prices = eval(input('Enter list of product prices: '))

#initializing a product counter with 0
product_count = 0

#iterating through each product price in the list
for price in prices:
    if price > 1000: #checking for products with price > 1000
        product_count += 1 #if true, incrementing the counter by 1
    
#printing the number of products satisfying the condition
print('Products above 1000:',product_count)
```

## Problem 5: Calculate Attendance Percentage

```python
#taking user input for a list of attendances
attendance = eval(input('Enter list of atttendances: '))

#initializing the Present count with 0
present = 0

#iterating through each attendance from the list
for i in attendance:
    if i == 'P':
        present += 1 #if present, increment the counter by 1

#calculating the attendance percentage
if len(attendance) > 0:
    percentage = (present/len(attendance))*100
else:
    percentage = 0.0 #handling the divide by zero error

#printing the attendance percentage
print('Attendance Percentage:',percentage)
```

## Problem 7: Count Character Frequency

```python
#taking user input for a string
string = input('Enter a string: ')

# creating an empty dictionary to store the characters and their occurrence count
string_dict = {}

#iterating through each character of the string
for s in string:
    if s in string_dict: # checking if the character is already present
        string_dict[s] += 1 #if yes, increment the frequency
    else:
        string_dict[s] = 1 #if no, append the character with frequency=1
        
#printing the dictionary with characters and their occurence count
print(string_dict)
```

## Problem 9: Check if a Key Exists in a Dictionary

```python
#creating an empty dictionary to store keys representing employee names and values representing employee salaries
employees = {}

#taking input until the user decides to stop
while True:
    # asking the user to enter an employee name
    name = input("Enter employee name (press Enter to stop): ") #if pressed enter, the loop will end
    
    if name == "":
        break # exit the loop if pressed enter

    # taking salary input for the entered employee
    salary = int(input("Enter salary: "))
    # storing the employee name and salary in the dictionary
    employees[name] = salary

#print(employees)

#taking user input for a employee name to check in dictionary
e_name = input('Enter an employee name to check: ')

#checking if the employee name exists in the dictionary
if e_name in employees.keys():
    print('Employee exists')
else:
    print('Employee doesnot exist')
```

## Problem 10: Calculate Average Marks

```python
#taking user input for a list of marks of a student
marks = eval(input('Enter a list of marks: '))

#initializing the sum as 0
total = 0

#iterating through each mark of the student from the marks list
for mark in marks:
    total += mark #adding each marks value to the total

#calculating the average marks
avg_marks = total/len(marks)

#printing the average marks
print('Average Marks:',avg_marks)
```
