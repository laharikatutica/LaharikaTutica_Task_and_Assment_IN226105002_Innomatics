## Problem 1: Social Media – Post Engagement Analyzer

```python
# Function to analyze engagement on a social media post
def like_counter(likes_list):
    # Variable to store the total number of likes
    total_likes = 0

    # Looping through each like count in the list
    for like in likes_list:
        # Add each like count to total_likes
        total_likes += like

    # Checking if the post is viral or not
    if total_likes >= 1000:
        status = "Viral Post"
    else:
        status = "Normal Engagement"

    # Displaying the results
    print(f"Total Likes: {total_likes}")
    print(f"Post Status: {status}")

# user input for number of like entries
number_of_entries = int(input("Enter number of like values: "))

user_likes = []

# Collecting like values from the user
for i in range(number_of_entries):
    like_value = int(input(f"Enter like count {i + 1}: "))
    user_likes.append(like_value)

# Calling the function with user-provided data
like_counter(user_likes)
```

## Problem 2: Healthcare – Medicine Stock Alert System

```python
# Function to check the availability status of medicine stock
def check_medicine_stock(stock_quantity):
    # If stock is less than 10, trigger a low stock alert
    if stock_quantity < 10:
        return stock_quantity, "Low Stock Alert"
    else:
        return stock_quantity, "Stock Sufficient"

# User input for the current medicine stock quantity
current_stock = int(input("Enter medicine stock quantity: "))

# Calling the function to check stock status
quantity, stock_status = check_medicine_stock(current_stock)

# Displaying the results
print("Medicine Stock:", quantity)
print("Status:", stock_status)
```

## Problem 3: Agriculture – Rainfall Adequacy Checker

```python
# Function to analyze seasonal rainfall data
def analyze_rainfall(rainfall_values, required_level):
    total_rainfall = 0

    # Looping to calculate total rainfall
    for value in rainfall_values:
        total_rainfall += value

    # Calculating the average rainfall
    average_rainfall = total_rainfall // len(rainfall_values)

    # Determining rainfall status
    if average_rainfall >= required_level:
        status = "Adequate Rainfall"
    else:
        status = "Inadequate Rainfall"

    # Displaying results
    print(f"Average Rainfall: {average_rainfall}")
    print(f"Rainfall Status: {status}")

# User input for number of rainfall readings
num_readings = int(input("Enter number of rainfall readings: "))

rainfall_data = []

# Collecting rainfall data from the user
for i in range(num_readings):
    rainfall = int(input(f"Enter rainfall value {i + 1}: "))
    rainfall_data.append(rainfall)

# User input for required rainfall level
required_rainfall = int(input("Enter required rainfall level: "))

# Calling the function
analyze_rainfall(rainfall_data, required_rainfall)
```

## Problem 4: Social Media – Duplicate Account Detection

```python
# Function to detect duplicate usernames in a system
def detect_duplicate_accounts(username_list):
    # Converting the list to a set to remove duplicates
    unique_usernames = set(username_list)

    # Comparing lengths to check for duplicates
    if len(username_list) != len(unique_usernames):
        return "Duplicate Accounts Found: Yes"
    else:
        return "Duplicate Accounts Found: No"

# user input for total number of users
total_users = int(input("Enter the total number of users: "))

usernames = []

# Collecting usernames from the user
for i in range(total_users):
    username = input(f"Enter username {i + 1}: ")
    usernames.append(username)

# Call the function and print the result
result = detect_duplicate_accounts(usernames)
print(result)
```

## Problem 6: Agriculture – Premium Crop Price Filter

```python
# Function to filter premium crop prices
def crop_price_filter(crop_price_list):
    premium_prices = []

    # Looping through each crop price
    for price in crop_price_list:
        # Checking if the price qualifies as a premium crop
        if price > 2000:
            premium_prices.append(price)

    return f"Premium Crops: {premium_prices}"


# user input for number of crop prices
number_of_crops = int(input("Enter the number of crop prices: "))

crop_prices = []

# Collecting crop prices from the user
for i in range(number_of_crops):
    price = int(input(f"Enter price of crop {i + 1}: "))
    crop_prices.append(price)

# Calling the function and print the result
result = crop_price_filter(crop_prices)
print(result)
```

## Problem 7: System Monitoring – Application Health Checker

```python
# Function to check application health based on error count
def application_health_checker(error_count):
    # Displaying the total number of errors
    print("Error Count:", error_count)

    # Determining system health status
    if error_count == 0:
        return "System Status: Healthy"

    elif error_count <= 5:
        return "System Status: Minor Issues"

    else:
        return "System Status: Critical Issues"


# user input for number of errors
number_of_errors = int(input("Enter the number of errors: "))

# Calling the function and print the result
system_status = application_health_checker(number_of_errors)
print(system_status)
```

## Problem 8: Banking – Daily Transaction Limit Checker

```python
# Function to check whether a transaction is within the daily limit
def check_daily_transaction_limit(transaction_amount):
    daily_limit = 50000  # Maximum allowed transaction per day

    # Displaying the transaction amount
    print("Transaction Amount:", transaction_amount)

    # Checking if transaction exceeds the daily limit
    if transaction_amount > daily_limit:
        return "Transaction Status: Rejected"
    else:
        return "Transaction Status: Approved"


# user input for transaction amount
amount = int(input("Enter your transaction amount: "))

# Calling the function and print the transaction status
transaction_status = check_daily_transaction_limit(amount)
print(transaction_status)
```

## Problem 9: E-Learning – Student Attendance Eligibility System

```python
# Function to check exam eligibility based on attendance
def check_attendance_eligibility(attendance_list):
    present_count = 0

    # Counting number of days present
    for status in attendance_list:
        if status == 'P':
            present_count += 1

    # Calculating attendance percentage
    attendance_percentage = (present_count / len(attendance_list)) * 100

    # Displaying attendance percentage
    print("Attendance Percentage:", attendance_percentage)

    # Checking eligibility
    if attendance_percentage >= 75:
        return "Exam Eligibility: Eligible"
    else:
        return "Exam Eligibility: Not Eligible"


# user input for total number of days
total_days = int(input("Enter the number of days: "))

attendance_records = []

# Collecting attendance data
for i in range(total_days):
    status = input(f"Enter the attendance for day {i + 1} (P/A): ").upper()
    attendance_records.append(status)

# Calling the function and print the result
eligibility_status = check_attendance_eligibility(attendance_records)
print(eligibility_status)
```

## Problem 10: Smart Electricity Bill Analyzer

```python
# Function to calculate electricity bill and usage level
def calculate_electricity_bill(units_consumed):
    # Calculating bill amount based on slab rates
    if units_consumed <= 100:
        bill_amount = units_consumed * 3
    elif units_consumed <= 200:
        bill_amount = (units_consumed - 100) * 5 + (100 * 3)
    else:
        bill_amount = (units_consumed - 200) * 7 + (100 * 5) + (100 * 3)

    # Classifying electricity usage based on bill amount
    if bill_amount < 500:
        usage_status = "Low Usage"
    elif bill_amount <= 1500:
        usage_status = "Moderate Usage"
    else:
        usage_status = "High Usage"

    # Returning both bill amount and usage status
    return bill_amount, usage_status


# user input for units consumed
units = int(input("Enter the number of units consumed: "))

# Calling the function
total_bill, usage_level = calculate_electricity_bill(units)

# Display results
print("Units Consumed:", units)
print("Total Bill:", total_bill)
print("Usage Level:", usage_level)
```
