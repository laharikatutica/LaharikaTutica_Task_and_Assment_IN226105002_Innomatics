## Problem 1: Smart Light Controller

```python
# Defining a class for a smart light device
class Smart_light:
    
    # Initializing the smart light and setting its initial status to OFF
    def __init__(self, name):
        self.name = name
        self.status = "OFF"

    # Turning the light ON
    def turn_on(self):
        self.status = "ON"

    # Turning the light OFF
    def turn_off(self):
        self.status = "OFF"

    # Displaying the current status of the light
    def display_status(self):
        print(f"{self.name} is {self.status}")

# user input for the name of the light and required action
light_name = input('Enter the light name: ')
action = input('Enter the Action (ON/OFF): ').upper()

# Creating an object
light = Smart_light(light_name)

# turning the light on
if action == 'ON':
    light.turn_on()
    light.display_status()

# turning the light off
elif action == 'OFF':
    light.turn_off()
    light.display_status()

# Handling invalid input
else:
    print('Invalid Action')
```

## Problem 2: Employee ID Card System

```python
# Defining a class for an employee ID card
class employee_id_card:
    
    # Initializing the employee ID card with name, ID, and department details
    def __init__(self, name, emp_id, department):
        self.name = name
        self.emp_id = emp_id
        self.department = department

    # Displaying all the employee ID card details
    def display_id_details(self):
        print('Employee ID Card')
        print(f"Name: {self.name}")
        print(f"ID: {self.emp_id}")
        print(f"Department: {self.department}")

# user input for employee details
name = input('Enter the Employee Name: ')
emp_id = input('Enter the Employee ID: ')
department = input('Enter the Department: ')

# Creating an object
id_card = employee_id_card(name, emp_id, department)

# Displaying the employee ID card info
id_card.display_id_details()
```

## Problem 3: Mobile Contact Record

```python
# Defining a class for a mobile contact
class Mobile_contact:
    
    # Initializing contact name and phone number
    def __init__(self, name, phone_no):
        self.name = name
        self.phone_no = phone_no

    # Displaying saved contact information
    def display_contact_info(self):
        print("Contact Saved")
        print(f"Name: {self.name}")
        print(f"Phone: {self.phone_no}")

# User input for contact details
name = input('Enter the contact name: ')
phone_no = int(input('Enter the phone number: '))

# Creating a contact object
contact = Mobile_contact(name, phone_no)

# Displaying the contact details
contact.display_contact_info()
```

## Problem 4: Product Price Tag Generator

```python
# Defining a class for product price
class Product_price:
    
    # Initializing product name and price
    def __init__(self, name, price):
        self.name = name
        self.price = price

    # Displaying product price details
    def display_price(self):
        print(f"Name: {self.name}")
        print(f"Price: ₹{self.price}")

# User Input for product details
name = input('Enter the product name: ')
price = int(input('Enter the product price: '))

# Creating a product price object
price_tag = Product_price(name, price)

# Showing the price details
price_tag.display_price()
```

## Problem 5: Movie Rating Display System

```python
# Defining a class for movie rating
class Movie_Rating:
    
    # Initializing movie name and rating
    def __init__(self, name, rating):
        self.name = name
        self.rating = rating

    # Displaying movie rating details
    def display_movie_rating(self):
        print(f"Movie: {self.name}")
        print(f"Rating: {self.rating} / 5")

# User input for movie name and its rating
name = input('Enter the Movie name: ')
rating = float(input('Enter the Movie Rating: '))

# Creating a movie rating object
movie_rating = Movie_Rating(name, rating)

# Showing the movie rating
movie_rating.display_movie_rating()
```

## Problem 6: Delivery Address Manager

```python
# Defining a class for delivery address
class Delivery_Address:
    
    # Initializing customer name and address
    def __init__(self, name, address):
        self.name = name
        self.address = address

    # Displaying delivery details
    def display_delivery_details(self):
        print('Delivery Details')
        print(f"Customer: {self.name}")
        print(f"Address: {self.address}")

# User Input for customer name and delivery address
name = input('Enter the Customer name: ')
address = input('Enter the Customer Address: ')

# Creating a delivery address object
customer_details = Delivery_Address(name, address)

# Showing delivery details
customer_details.display_delivery_details()
```
