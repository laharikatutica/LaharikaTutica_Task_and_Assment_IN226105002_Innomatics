## Problem 1: Unique Words in a Sentence

```python
#taking a user input sentence
sentence = input('Enter sentence: ')

#splitting the sentence into a list of words
words = sentence.split()

#creating an empty list to store the unique words
unique_words = []

#unique word counter initialised with 0
w_count = 0

#iterating through each word
for word in words:
    #checking for uniqueness of the word
    if word not in unique_words:
        unique_words.append(word) #if unique, append to the unique_words list
        w_count += 1 #incrementing the unique word counter by 1
    else:
        pass

#printing the Unique words count and the unique words as a list
print('Unique words count:',w_count)
print('Unique words:',unique_words)
```

## Problem 2: Highest Salary from Employee Data

```python
#given a dictionary with keys representing employee names and values representing employee salaries
employees = {'Ravi':75000,'Anita':68000,'Kiran':72000}

#assuming the first employee salary as the highest
highest_name,highest_salary = next(iter(employees.items()))

#iterating through each employee salary to get the highest salary
for name,salary in employees.items():
    if salary > highest_salary:
        highest_salary = salary
        highest_name = name

#printing the highest salary with the employee name
print(f'Highest Salary: {highest_name} - {highest_salary}')
```

## Problem 3: Find Maximum and Minimum Values

```python
#given list of integers
numbers = [45, 22, 89, 10, 66]

#assuming the minimum and maximum to be the first integer vsalues of the list
min_num = numbers[0]
max_num = numbers[0]

#iterating through each number of the list
for num in numbers:
    #checking for minimum number
    if num < min_num:
        min_num = num
    
    #checking for maximum number
    if num > max_num:
        max_num = num
        
#printing the minimum and maximum numbers from the list and the list itself
print('List:',numbers)
print('Max:',max_num)
print('Min:',min_num)
```

## Problem 4: Count Products Above a Price Threshold

```python
#given a list of prices of the products
prices = [450, 1200, 899, 1500, 300]

#initializing a product counter with 0
pro_count = 0

#iterating through each product price in the list
for price in prices:
    if price > 1000: #checking for products with price > 1000
        pro_count += 1 #if true, incrementing the counter by 1
    
#printing the number of products satisfying the condition
print('Products above 1000:',pro_count)
```

## Problem 5: Calculate Attendance Percentage

```python
#given a list of attendances
attendance = ["P", "P", "A", "P", "P"]

#initializing the Present count with 0
present = 0

#iterating through each attendance from the list
for i in attendance:
    if i == 'P':
        present += 1 #if present, increment the counter by 1

#calculating the attendance percentage
if len(attendance) > 0:
    percentage = (present/len(attendance))*100
else:
    percentage = 0.0 #handling the divide by zero error

#printing the attendance percentage
print('Attendance Percentage:',percentage)
```
