### 1. User Login Check

```python
# given username and password
username = "admin"
password = "1234"

# User-input for username and password
login_user = input('Enter your Username: ')
login_password = input('Enter your Password: ')

#matching the entered username and password with the given
if login_user == username and login_password == password:
    print('Login Successful')
else:
    print('Invalid Credentials')
```

### 2. Pass / Fail Analyzer

```python
#given marks of multiple students
marks = [45, 78, 90, 33, 60]

#initializing a counter for passed students with 0
p_count = 0

#initializing a counter for failed students with 0
f_count = 0

#iterating though each student's marks
for m in marks:
    if m >= 50: #passing condition
        p_count += 1 #incrementing the passed students count by 1
    else:
        f_count += 1 #incrementing the failed students count by 1
        
#printing final result
print('Number of passed students:',p_count)
print('Number of failed students:',f_count)
```

### 3. Simple Data Cleaner

```python
#given list of user names
names = [" Alice ", "bob", " CHARLIE "]

#creting an empty list to store the cleaned names
cleaned_names = []

#iterating through each name
for name in names:
    name = name.strip().lower()
    #strip for removing extra spaces from both sides of the name
    #lower for converting the name to lowercase
    cleaned_names.append(name) #appending it to the new list
    
#printing the list of cleaned names
print('The cleaned list of names:',cleaned_names)
```

### 4. Message Length Analyzer

```python
#given list of messages
messages = ["Hi", "Welcome to the platform", "OK"]

#iterating through each message
for message in messages:
    print(f"Message:{message}") #printing each message
    print(f"Length:{len(message)}") #printing length of each message
    
    #checking if length > 10 for each message
    if len(message) > 10:
        print("Message flagged: Length exceeds 10 characters") #flagging messages longer than 10 characters
    print() #printing blank line for cleaner display
```

### 5. Error Message Detector

```python
#given list of system logs
logs = ["INFO", "ERROR", "WARNING", "ERROR"]

#initializing an error counter with 0
e_count = 0

#iterating through each log entry
for log in logs:
    if log == 'ERROR': 
        e_count += 1 #incrementing the error count by 1
        
#printing the total error count
print('Total Error Count:',e_count)
```
